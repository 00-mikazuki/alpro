```python
print( '''  
menu
1. penjumlahan
2. pengurangan
''')

stop=False

while not(stop):
    menu=int(input('pilih menu: '))
    a1=int(input('masukkan angka 1 = '))
    a2=int(input('masukkan angka 2 = '))
    
    if menu==1:
        penj=a1+a2
        print('hasil penjumlahan dari',a1,'+',a2,'=',penj)
    elif menu==2:
        peng=a1-a2
        print('hasil pengurangan dari',a1,'-',a2,'=',peng)
    
    lanjut=str(input('lanjut y/t = '))
    
    if lanjut=='t':
        stop=True
```
